### Seudocódigo
* Entrada -> grafo (nodos, aristas, pesos)
* Elegir del grafo un origen y un destino
* Categorizar todos los nodos como no visitados, excepto el nodo de origen
* El nodo de origen se vuelve permanente
* Seleccionar el nodo permanente, escanear los nodos directamente conectados y evaluar el peso de moverse en cada uno
* Todos los nodos se vuelven temporales
* Se selecciona el nodo con menor peso y se vuelve permanente
* Se repite el proceso con ese ultimo nodo permanente
* El algoritmo termina cuando no existan nodos termporales

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
Grafo = nx.Graph()
Grafo
```

```python
Grafo.add_edge("Casa", "Parque", weight = 2)
Grafo.add_edge("Casa", "Tienda", weight = 5)
Grafo.add_edge("Parque", "Tienda", weight = 1)
Grafo.add_edge("Parque", "Escuela", weight = 4)
Grafo.add_edge("Tienda", "Cine", weight = 3)
Grafo.add_edge("Escuela", "Cine", weight = 1)

inicio = "Casa"
destino = "Cine"
```

```python
ruta_mas_corta = nx.dijkstra_path(Grafo, source=inicio, target=destino, weight="weight")
```

```python
longitud_ruta = nx.dijkstra_path_length(Grafo, source=inicio, target=destino, weight="weight")
```

```python
print(ruta_mas_corta)
print(longitud_ruta)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(Grafo)

nx.draw(Grafo, pos, with_labels = True)
```
