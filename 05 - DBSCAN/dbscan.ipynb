### Seudocódigo
* Entrada -> nube de puntos, épsilon, min_pts
* Para cada punto NO visitado de la nube de puntos:
* Marcar el punto como visitado
* Dibujar el epsilon
* Contar la cantidad de puntos que están dentro de épsilon
* Si la cantidad de puntos es mayor o igual a min_pts -> Punto corazón
* Si la cantidad de puntos es menor a min_pts, pero el punto hace parte de un cluster -> punto frontera
* Si la cantidad de puntos es menor a min_pts, y el punto no hace parte de un cluster -> ruido
* Termina cuando todos los puntos se hayan marcado como visitados

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples = 300, noise = 0.2, random_state = 42)

dbscan = DBSCAN(eps = 0.2, min_samples = 15)

cluster_predichos = dbscan.fit_predict(X)

plt.figure(figsize = (10, 6))
scatter = plt.scatter(X[:, 0], X[:, 1], c = cluster_predichos, cmap = "plasma")
plt.title("DBSCAN Clustering")
```
