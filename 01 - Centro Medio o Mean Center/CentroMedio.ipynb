### Pseudocódigo
* Definir variables, suma_total_x, suma_total_y, cantidad_amigos

* Recorremos cada amigo y de cada uno haremos lo siguiente:
* Del amigo sumamos la coordenada X a suma_total_x
* Del amigo sumamos la coordenada y a suma_total_y

* Dividir la suma_total_x en cantidad_amigos
* Dividir la suma_total_y en cantidad_amigos
* Obtener el punto centro como el lugar donde tomaremos cerveza

```python
casas_amigos = [
    (4.6097, -74.0817),
    (4.6250, -74.0650),
    (4.6400, -74.0900),
    (4.5900, -74.1200),
    (4.7100, -74.0700)
]

suma_total_x = 0
suma_total_y = 0
cantidad_amigos = len(casas_amigos)

for amigo in casas_amigos:
    x = amigo[0]
    y = amigo[1]

    suma_total_x += x
    suma_total_y += y

centro_x = suma_total_x / cantidad_amigos
centro_y = suma_total_y / cantidad_amigos

print(centro_x, centro_y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = {
    "Amigo": ["Juan", "Ana", "Pedro", "Carolina", "Beto"],
    "Latitud": [4.6097, 4.6250, 4.6400, 4.5900, 4.7100],
    "Longitud": [-74.0817, -74.0650, -74.0900, -74.1200, -74.0700]
}

df = pd.DataFrame(datos)
print(df)
```

```python
centro_medio = df[["Latitud", "Longitud"]].mean()
print(centro_medio)
```

```python
plt.scatter(df["Longitud"], df["Latitud"], color="blue", label="Amigos")
plt.scatter(centro_medio["Longitud"], centro_medio["Latitud"], color="red", marker="*", label="Cerveza")
plt.legend()
plt.title("Lugar para tomar Cerveza")
```
