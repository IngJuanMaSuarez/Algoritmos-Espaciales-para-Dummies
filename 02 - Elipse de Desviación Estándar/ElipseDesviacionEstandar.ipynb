### Pseudocódigo
* Obtener el centro medio
* Obtener la matriz de varianzas covarianzas 
* (calcular la distancia de cada coordenada con el promedio al cuadrado)
* Analizar como cambia la coordenada X con la coordenada Y basado en la covarianza 
* Desviaciones estandar del eje mayor
* Desviaciones estandar del eje menor
* obtener la elipse centrada en el centro medio y rotada por el angulo calculado

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

np. random.seed(42)
x = np.random.normal(scale = 2, size = 50) + np.linspace(0, 10, 50)
y = x +np.random.normal(scale = 1, size = 50)
puntos = np.column_stack((x, y))

centro_x = np.mean(x)
centro_y = np.mean(y)

covarianza = np.cov(x, y)

val_propios, vec_propios = np.linalg.eig(covarianza)

angulo = np.degrees(np.arctan2(vec_propios[1, 0], vec_propios[0, 0]))

ancho = 2 * np.sqrt(val_propios[0])
alto = 2 * np.sqrt(val_propios[1])
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.scatter(x, y, label = "Accidentes")
ax.scatter(centro_x, centro_y, color = "red", marker = "*", s = 150, label = "Centro Medio")
elipse = Ellipse(xy=(centro_x, centro_y), width = ancho, height = alto, angle = angulo, edgecolor = "red", fc = "None", lw = 2, linestyle = "--")
ax.add_patch(elipse)
plt.legend()
plt.show()
```
